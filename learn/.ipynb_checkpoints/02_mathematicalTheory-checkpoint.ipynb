A soundwave's graphical representation is the representation of the movement of air molecules over time.
- X-axis = time
- Y-axis = how much it moves (= Amplitude or "how loud ?")

Of course the movement on the Y-axis can take a certain amount of time:<br>
- The *Period (T)* is the time required for one cycle (in seconds).
- The *Frequency (f)* is the number of periods per second in (Hz = cycle/second).

In other words:
$$T = \frac{1}{f}$$$$f = \frac{1}{T}$$

And as we all know, a soundwave can be represented by a sine. In numpy, a sine is a `np.sin()` function. It calculates the sine value of an angle.

But what is the sine value of an angle α?<br>
In a right trangle, the sine of an accute angle α is the relation between its opposite side over the hypotenuse of the triangle.
$$\sin{α} = \frac{opposite}{hypotenuse}$$
In the `np.sin()` function, the angle α is expressed in ***Radiants***. One radiant is a standard unit of measure so that $2\pi * rad = 360°$.

### Graphing soundwaves with `np.sin()`

````python
ipd.Image("https://upload.wikimedia.org/wikipedia/commons/3/3b/Circle_cos_sin.gif")
````

To create a graphical representation of a soundwave using `np.sin()` we need to input an angle (in radiants) in the function. However, when dealing with soundwaves, what we usually know is its frequency, amplitude and the time elapsed...<br>
But it is enough :-D

Since:
- the period (T) is the time (seconds) required for one cycle
- the frequency (f) is the amount of cycles per seconds
- one cycle is equal to $360°$
- $360° = 2\pi * rad$

We can deduce that, for a period T :  $cycle = 360° = 2\pi * \frac{cycles}{second} * T$<br>
This means that for a time t: $angle = 2\pi * \frac{cycles}{second} * t$<br>
And, thus since the radiant is an angle:
$$rad = 2\pi * f * t$$

````python
f = 0.5 # Frequency of 0,5 Hz
time = np.linspace(0, 6, 60000)
amplitude = np.sin(2 * np.pi * f * time)
plt.figure(figsize = (10, 4))
plt.plot(time, amplitude)
plt.title("Soundwave of Frequency of 0.5 Hz")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True, which='both')
plt.show()
````

### Explaining the Fourrier Transform
If we apply a Fourrier Transform (`scipy.fft.fft(d, l)` with scipy or `np.fft.fft(d, l)` with numpy) to an array (`d`) containing the sinusoidal signal of a given frequency of `l` samples, we will obtain the following:

$$\begin{bmatrix}1 \\ f \\ 1\end{bmatrix} \overrightarrow{FFT} \begin{bmatrix}1 \\ \hat{f} \\ 1\end{bmatrix}$$

$\hat{f}$ is an array of **complex numbers** (*with a magnitude and a phase*):
- the magnitude tells us how strong that frequency is
- the phase tells us the 'mixture' of the signal: is it rather a cosine or a sine ?

This means that each value will be composed of both ***a real and imaginary part*** such as each value is equal to "$x + yi$" where `x` and `y` are real numbers and `i` $= \sqrt{-1}$.<br>
Also, we call ***conjugate*** the complex number with an equal real part and an imaginary part *equal in magnitude but opposit in sign*.<br>
> $\lambda = x + iy$<br>
> $\bar{\lambda} = x - iy$

````python
ipd.Image(
    url="https://upload.wikimedia.org/wikipedia/commons/6/69/Complex_conjugate_picture.svg",
    width=150,
    height=300
)
````

Since what we want to obtain is the ***magnitude*** of each frequency, here is a simple solution:

Let's start with $\lambda$ such as $\lambda$ is one item within the array of Fourrier Transformed points:<br>
> $\lambda \bar{\lambda} = (x + iy)(x - iy)$<br>
> $\lambda \bar{\lambda} = x^2 - i^2y^2$

But since $i = \sqrt{-1}$, this means $i^2 = -1$.

> $\lambda \bar{\lambda} = x^2 - (-1)y^2$<br>
> $\lambda \bar{\lambda} = x^2 + y^2$

This will give us only positive numbers and will exponentially increase the visibility of any given value.


If we observe the resulting graph, we will see that the second half of the x-axis is the mirror image of the first half.
This is due to the fact that we are working with:
1. a circle (360°)
2. exponants (to the 2nd degree) which is always positive

In other words, we only need the first half of the graph to have useful values.

### Putting FFT and Frequencies in relation
Ok, so we now have a graph that shows the **magnitude** of an item over its **item number**.<br>
Since we know the sample rate of the sound, we can infer the time at which the magnitude occurs.

If `n = len(timeInSeconds * sampleRate)` and `sp = timeInSecondsBetweenEachSample` then:

$$\frac{1}{sp * n} * |\lambda|² = \frac{|\lambda|²}{n * sp} = frequency$$

Where the frequency is x/second.<br>
This means that if we take observe the `np.floor(n/2)` first elements of the results of the FFT $\lambda$ over the `np.floor(n/2)` first elements of the frequency axis we have just built (since we only need the first half).
We will obtain a graph showing us all the frequencies present in a sound wave.

### Demonstration

````python
from scipy.fft import fft, ifft
````

````python
# VARIABLES
f0 = 50 # Frequency (in Hz)
f1 = 125
d = 0.5 # duration (in seconds)
sr = 1000 # number of sample points per second (= sample rate)
sp = 1/sr
````

````python
# TIME AND SINE-WAVES
time = np.arange(0, d, sp)
y1 = np.sin(2 * np.pi * f0 * time)
y2 = np.sin(2 * np.pi * f1 * time)

multiple = y1 + y2
````

````python
# Plot Sinusoidal
plt.figure(figsize = (14, 5))
plt.plot(time, y1, label="50 Hz", linewidth=3)
plt.plot(time, multiple, label="Sum")
plt.title("Sinusoidal signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.xlim(time[0], time[-1])
plt.grid()
plt.legend()
plt.show()
````

````python
# Plot Magnitude values obtained through the Fourrier Transform
four_y = fft(multiple, len(time)) # Compute FFT with args: data and length of data

plt.figure(figsize = (14, 5))
plt.plot(four_y.real)
plt.title("Fourrier Transform of the Y-axis of the sinusoidal")
plt.xlabel("Point number")
plt.ylabel("Absolute Value of Fourrier Transform")
plt.grid()
````

````python
# Plot Spectral analysis

# Step 1: generate a frequency axis
n = len(time)
freq_axis = np.arange(n) / (sp * n)

# Step 2: generate the Power Spectrum (Magnitude per frequency)
psd = four_y * np.conj(four_y)

# Step 3: We only need the first half
first_half = np.arange(1, np.floor(n/2), dtype='int') # saved as an array

# Step 4: Plotting
plt.figure(figsize = (14, 5))
plt.plot(freq_axis[first_half], psd[first_half].real)
plt.title("Power Spectrum vs. Frequency (Hz)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude of frequency")
plt.grid()
````

As we can see on the graph above, we have a clear peak at a frequency of 50 Hz and 125 Hz. These peaks correspond to the frequencies we have chosen for `f0` and `f1` in our example. We summed the two frequencies up to generate a sum of sine waves and ... TADAAA we managed to exctract frequencies from a sound.

### Real World application

In the real world, this would of course be a bit more complicated. We need to take into consideration the noise can be produced by the electromagnetic interference of the cables, or simply noise present on the sound source because it has been recorded in bad quality.

To do this we can apply a filter.

To show how a filter works, we will take the same frequencies as before and will add noise of random frequency to each sample using the [`np.randn`](https://numpy.org/devdocs/reference/random/generated/numpy.random.randn.html) method.
```python
y_clean = y1 + y2
noise = np.random.randn(len(time)) * 2.5  # 2.5 is arbitrary value I chose
y_noizy = y_clean + noise
```

After applying the FFT to the dataset and generating its Power Spectrum, we can see that it is noizy but that it still has very clean peaks at the 50 Hz and 125 Hz marks.

This means that we can ***cut off*** all the values below a certain magnitude to get rid of the noise.

This is done through a filter. The filter will simply remove (or render 0) all values where $$|\hat{\lambda}| < magnitude$$

Just for the purpose of this tutorial, we will also convert the "cleaned up" Power spectrum back to a sum of sine waves with the **inverse Fourier Transform** or `ifft()` method.

*sources:*
- *[Denoizing video](https://www.youtube.com/watch?v=s2K1JfNR7Sc)*
- *[Fourier Transform](https://www.youtube.com/watch?v=r6sGWTCMz2k)*
- *[`scipy.fft.fft`](https://docs.scipy.org/doc/scipy-1.6.0/reference/generated/scipy.fft.fft.html#scipy.fft.fft)*
- *[Useful Math](https://www.symbolab.com/solver?origin=ddg&query=%28x+%2B+y%29%28x+-+y%29)*

````python
# Creating noise
y_clean = y1 + y2
noise = np.random.randn(len(time)) * 2.5  # 2.5 is arbitrary value I chose
y_noizy = y_clean + noise
````

````python
# Filtering

# Step 1: Choose a filter value
filter_value = 100

# Step 2: Filter out the values
indices = psd > filter_value # Find all the frequencies with a power larger to the filter value
clean_psd = psd * indices    # Zero out all values where indices == False
four_y = four_y * indices    # Zero out the smallest Fourier coeffs in Y

# Step 3: Invers Fourier Transform
filtered_fourier = ifft(four_y)
````

````python
# Plotting the lot
fig, axs = plt.subplots(3, 1, figsize=(14.5, 10))

plt.sca(axs[0])
plt.plot(time, y_clean, label="Clean", linewidth=3)
plt.plot(time, y_noizy, label="Noizy", linewidth=0.8)
plt.title("Base signals")
plt.xlim(time[0], time[-1])
plt.legend()

plt.sca(axs[1])
plt.plot(time, filtered_fourier.real, label="Filtered")
plt.title("Filtered signal")
plt.xlim(time[0], time[-1])
plt.ylim(-10, 10)
plt.legend()

plt.sca(axs[2])
plt.plot(freq_axis[first_half], (psd[first_half] / n).real, label="Noizy")
plt.plot(freq_axis[first_half], (clean_psd[first_half] / n).real, label="Clean")
plt.xlim(freq_axis[first_half[0]], freq_axis[first_half[-1]])
plt.legend()

plt.show()
````

## Dealing with Harmonics

In music, a frequency `f` is never pure. I is always accompanied by harmonics due to the limited length of the string (for string instruments including piano) or of the vibrating surface in general.

So what is an harmonic in terms of frequency ?

> A harmonic is the sequence of frequencies, in which each frequency is an integer multiple of a fundamental.<br>
*[Wikipedia](https://en.wikipedia.org/wiki/Harmonic_series_(music))*

````python
ipd.Image("https://upload.wikimedia.org/wikipedia/commons/thumb/c/c5/Harmonic_partials_on_strings.svg/620px-Harmonic_partials_on_strings.svg.png", height=300, width=320)
````

This means that for any frequency `f0`, its harmonics are running at a frequency of $n * f0$.<br>
The problem is that the harmonics are never "as loud" as the fundamental.

Check [this video](https://www.youtube.com/watch?v=0Rfushlee0U) for more info.

To visualize the problem let's picture a string of an instrument being plucked.<br>
- The string vibrates at a frequency of `n * f` (where $n = 1$) with an amplitude of `a`
- Each time the wave bounces of the limit of the string, the amplitude is divided by 2
- Each time the amplitude is divided by 2, an additional harmonic is created multiplying f by $n + 1$.


With this, we simplify truth a lot. But it allows us to create the formula:
$$freq = \sum_{i = 1}^{n} \frac{1}{n} * \sin(2\pi * f * n * time)$$

````python
harmonics = [1, 2, 3, 4, 5]
f0 = 110
d = 5
sr = 1000

t = np.arange(0, d, 1/sr)
y0 = np.sin(2 * np.pi * f0 * t)
y1 = (1/2) * np.sin(2 * np.pi * f0 * t * 2)
y2 = (1/3) * np.sin(2 * np.pi * f0 * t * 3)
y3 = (1/4) * np.sin(2 * np.pi * f0 * t * 4)

y = sum([y0, y1])

plt.plot(t, y0)
plt.plot(t, y1)
plt.plot(t, y2)
plt.plot(t, y3)
plt.plot(t, y)

plt.title("Sinusoidal signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.xlim(t[0], t[-1])
plt.grid()
plt.show()
````

````python
n = len(time)
fy = fft(y, n)

freq_axis = np.arange(n) / ((1/sr) * n)
psd = fy * np.conj(fy)

half = np.arange(1, np.floor(n/2), dtype='int') # saved as an array

# Step 4: Plotting
plt.figure(figsize = (14, 5))
plt.plot(freq_axis[half], psd[half].real)
plt.title("Power Spectrum vs. Frequency (Hz)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude of frequency")
plt.grid()
````
