# Music: Tuning & Interval visualization

Since we aim to recognize notes to generate scores, we first need to have a clear understanding of the science between intervals and the tuning systems that we have at our disposal.

## Tuning systems
### Twelve-tone equal temperament
The 12-tone equal tmeperament (or TET) divides an octave in 12 equaly divided semitones. Each semitone has an interval of 100 ***cents***. This means that an octave has 1200 cents.

In frequency, each semitone (100 cents) is at an interval equalling $2^{1/12}$ (â‰ˆ 1.05946) of the previous semitone.

> *Example:<br>
> A440 = 440 Hz<br>
> A# = 440 * $2^{1/12}$ = 466,1638 Hz<br>
> B = 440 * $2^{2/12}$ = 493,8833 Hz*



### Just intonation
Just intonation is the fact of using what is called ***just intervals***. This means that the frequency of a specific interval will not be calculated based on the total amount of "space" between the root and its octave (like in TET) but on the note itself.

We know that when you double the frequency of a note, you get the octave of that note. If you multiply by 1.5, you get the perfect fifth.

| Note | Frequency | Multiplier | In fraction |
| --- | --- | --- | --- | 
| A | 440 Hz | 1.0 | 1/1 |
| C# | 550 Hz | 1.25 | 5/4 |
| E | 660 Hz | 1.5 | 3/2 |
| A<sup>+1</sup> | 880 Hz | 2.0 | 2/1 |
| C#<sup>+1</sup> | 1100 Hz | 2.5 | 10/4 |
| E<sup>+1</sup> | 1320 Hz | 3.0 | 6/2 |

This intonation is great for a single scale, but over a tune, this might create a change in pitch.<br>
If we were to use the TET as basis, the interval between A and C# would be calculated at 386 cents instead of the expected 400 cents.<br>
This means that if we were to keep C# as root and calculate the minor 6th (A | minor 6th = 128/81 for 3-limit tuning) we would have 869.14 Hz. This is 7.82 cents lower than the expected 880 Hz.

## Analysis

Global variables

```python
# IMPORTS
%matplotlib inline
import scipy
import scipy.signal
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd


plt.rcParams["figure.figsize"] = (13, 5)


# VARIABLES
d = 5                 # duration (in seconds)
sr = 44100               # sample rate (Hz)
fcutoff = 4000  # frequency of the cutoff filter
```

```python
def createTone(f0, harmonics=[1]):
    """Returns a tone of frequency 'f0' with his specified harmonic amplitudes.
    - For a frequency (f) over time (t) the sine wave of the frequency is:
        np.sin(2 * np.pi * f0 * t)
    - The harmonic of a frequency is a multiple of the frequency (up or down).
    - Calculating a sine of a harmonic (h) of frequency (f * m) over time (t) is:
        1/m * sin(2*pi * f * t * m)
    """
    time = np.arange(0, d, 1/sr)
    y = None
    for i, h in enumerate(harmonics, 1):
        sine = h * np.sin(2 * np.pi * f0 * time * i)
        if y is None:
            y = sine
        else:
            y += sine
    return y
```

```python
def filterTone(x, fcutoff=None):
    """Low-pass filtered signal.
    """
    if fcutoff is None:
        fcutoff = sr / 2.0
    
    h = scipy.signal.firwin(
        55,  # numtaps
        2 * float(fcutoff) / sr  # cutoff
    )
    lowpass = np.convolve(x, h)
    return lowpass
```

```python
def makeDoubleStop(ratio, f0=440, harmonics=[1]):
    """Create 2 tones that will be played simultaneously and plot the tones'
    spectra.
    """
    f1 = ratio * f0
    
    # Generate both tones
    x0 = filterTone(createTone(f0, harmonics), fcutoff)
    x1 = filterTone(createTone(f1, harmonics), fcutoff)
    
    # Sum up both tones and normalize
    y = x0 + x1
    y = 0.5 * (y / y.max())  # Amplitude on 1 instead of 2 without changing f
    
    # Fourier Transform
    fx0 = scipy.fft.fft(x0)
    fx1 = scipy.fft.fft(x1)
    fy = scipy.fft.fft(y)
    
    # Create the frequency-axis
    n = len(fx0)
    freq = np.arange(n) / ((1/sr) * n)
    
    # Plot spectrum for both notes
    plt.semilogy(freq, abs(fx0), marker='o')
    plt.semilogy(freq, abs(fx1), color='r')
    plt.xlim(xmin=0, xmax=(len(harmonics)+1)*(f1 if ratio > 1 else f0))
    plt.ylim(ymin=0.1)
    plt.xlabel("Frequency (Hz)")
    plt.legend(('f0 = %.2f Hz' % f0, 'f0 = %.2f Hz' % f1))
    
    # Output audio widget.
    print('ratio between notes:', ratio)
    print('difference in cents:', np.log2(ratio)*1200)
    return ipd.Audio(y, rate=sr)
```

### Unison (2x the same note)

```python
makeDoubleStop(1)
```

### Octave (Ratio of 2)

```python
makeDoubleStop(2, harmonics=[1, 0.5])
```

#### Calculating the Ratio and the Harmonics:

The ratio is equal to the ratio of the frequency compared to `f0`.

This means that if $f1 = 2 * f0$, the ratio is `2`.<br>
By extension, any interval within the octave would be equal to $f1 = x * f0$ so that $1 \lt x \lt 2$.

> **IF:**<br>
> $f1 = 2 * f0$<br>
> $f1 = 2^1 * f0$<br>
> $f1 = 2^\frac{12}{12} * f0$<br>
> **THEN:**<br>
> $f = 2^\frac{x}{12} * f0$



So, in Equal Temperament, since one octave is 1200 cents it would be equal to:
| note | difference in Cents | ratio |
| --- | --- | --- |
| Root | 0 | 1 |
| Second | 200 | `2**(2/12)` |
| Third | 400 | `2**(4/12)` |
| Fourth | 500 | `2**(5/12)` |
| Fifth | 700 | `2**(7/12)` |
| Sixth | 900 | `2**(9/12)` |
| Seventh | 1100 | `2**(11/12)` |

We can create a formula to calculate that for us:

```python
def ratio(cents):
    c = cents / 100
    return float(2**(c/12))
```
