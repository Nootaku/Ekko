# Reading Audio Files

My first approach to audio files is to read the RAW data of a file. Without it, we cannot do anything.

The most logical (and scientific) way to do this is to read the binary data of the audio file. This makes it possible to easily create a list by reading the file as a binary object and converting it to a NumPy array.

Let's go with the basic approach of what we already know:

````python
import os
import numpy as np

# Path to audio files
a_path = "audio/440khz.wav" # one second of tone at 440kHz
song_path = "audio/doremi.wav" # Audio file where a piano played C3, D3, E3
````

#### The `with open( )` method

````python
with open(a_path, "rb") as file:
    raw_audio = np.frombuffer(file.read())

print(raw_audio)
````

#### Using Librosa
Documentation regarding the Librosa module can be found [here](https://librosa.org/doc/latest/index.html).

The one thing we are going to focus on is how to read an Audio File. Following the documentation we see the following:
```python
y, sr = librosa.load(filename)
```
- ***y***: time series = One-dimensional NumPy Array
- ***sr***: sampling rate (of y)

This means that librosa does the dirty work for us:<br>The conversion of binary to numpy array is already done

````python
import librosa
````

````python
y, sr = librosa.load(a_path, sr=None)
print(y)
print(sr)
````

Since the `librosa` allows much easier access to data, we will use that method from now on.

We need to figure out one problem, though: why are the values different ?

# Visualizing Audio

Now that we can read audio, it is only normal that we try to visualize it. I had previous experience with `matplotlib` so I want to try.

````python
import matplotlib.pyplot as plt

cde, samp = librosa.load(song_path, sr=None)
````

````python
time = np.linspace(0, len(cde) / samp, num=len(cde))

plt.figure(figsize=(14, 5))

plt.title("File Graph")
plt.plot(time, cde, color="blue")
plt.xlabel("Time (sec)")
plt.ylabel("Amplitude")

plt.show()
````

But librosa has a method that can be used as well: `librosa.display.waveplot`.

````python
import librosa.display

plt.figure(figsize=(14, 5))
librosa.display.waveplot(cde, sr=samp)

plt.show()
````

### Spectrogram ?
Librosa has also an inbuild spectrogram:<br>
`librosa.display.specshow`

````python
X = librosa.stft(cde)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=samp, x_axis='time', y_axis='hz')

plt.show()
````

### Quick tutorial on Playing and Writing Audio

In order to play and write audio, we can use Jupyter's inbuild Audio module:<br>
`Ipython.display.Audio`

To write audio Librosa uses PySoundFile (as [documented here](https://librosa.org/doc/0.8.0/ioformats.html?highlight=write#write-out-audio-files)). The goto function is:<br>
`soundfile.write(path, data, sample_rate)`

````python
import IPython.display as ipd
import soundfile
````

````python
# Write file
e_path = "audio/tone_329-63.wav"
sample_rate = 44100
time = 2.5 # in seconds
t = np.linspace(0, time, int(time * sample_rate), endpoint=False)
x = 0.5 * np.sin(2 * np.pi * 329.63 * t)

soundfile.write(e_path, x, sample_rate)

# Read local file
ipd.Audio(e_path)
````
